```python
import os
import cv2 as cv
import numpy as np

def find(img, x, y):
    x0 = x
    xf = x
    y0 = y
    yf = y
    queue = [
    (x + 1, y),
    (x - 1, y),
    (x, y + 1),
    (x, y - 1)
    ]
    while len(queue) > 0:
        x, y = queue.pop()

        if y < 0 or y >= len(img):
            continue
        row = img[y]

        if x < 0 or x >= len(row):
            continue
        pixel = row[x]

        if pixel == 255:
            continue
        img[y][x] = 255

        x0 = min(x0, x)
        xf = max(xf, x)
        y0 = min(y0, y)
        yf = max(yf, y)

        queue.append((x + 1, y))
        queue.append((x - 1, y))
        queue.append((x, y + 1))
        queue.append((x, y - 1))

    return ((x0, y0), (xf, yf))

def Segmentation(image):

    org = cv.imread(image)
    img = org.copy()

    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    threshold, img = cv.threshold(
        img, 110, 255, cv.THRESH_BINARY
    )

    if img[0,0] < 120:
        img = cv.bitwise_not(img)
        
    rects = []
    for i in range(len(img)):
        row = img[i]
        for k in range(len(row)):
            if row[k] == 0:
                # print("Letra")
                rects.append(find(img, k, i))
                
                # img_teste = img[crop[0][1] : crop[1][1], crop[0][0] : crop[1][0]]
    for i in rects:
        print(i)
        
    mark = org.copy()
    img_teste = []

    for rect in rects:
        mark = cv.rectangle(mark, rect[0], rect[1], (0, 255, 0), 2)
        cropped_img = org[rect[0][1]:rect[1][1], rect[0][0]:rect[1][0]].copy()
        img_teste.append(cropped_img)

    output_folder = "Words"

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for i, cropped_img in enumerate(img_teste):
        output_path = os.path.join(output_folder, f"crop_{i}.png")
        cv.imwrite(output_path, cropped_img)

```
